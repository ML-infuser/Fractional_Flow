```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_excel('data.xlsx')
df
```

```python
X = df.iloc[3:,:4]
X
```

```python
#CASE 1
```

```python
krkw1 = np.array(X['Unnamed: 3'][1:]).astype(float)
mwmo1 = 0.01
krkw1
```

```python
n = len(krkw1)
fw1 = []
for i in range(n):
  x = 1/(1+krkw1[i]*mwmo1)
  fw1.append(x)
```

```python
fw1 = np.array(fw1)
fw1
```
